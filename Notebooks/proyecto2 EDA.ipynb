```python
import numpy as np
from tensorflow.keras.datasets import imdb
import matplotlib.pyplot as plt
import seaborn as sns
# Cargar el dataset
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)
# Tamao de las reseas
review_lengths = [len(review) for review in X_train]
# Histograma de longitudes de reseas
plt.figure(figsize=(10,6))
sns.histplot(review_lengths, bins=50, kde=True)
plt.title("Distribucin de la longitud de las reseas")
plt.xlabel("Longitud de la resea")
plt.ylabel("Frecuencia")
plt.show()
# Distribucin de clases
sns.countplot(x=y_train)
plt.title("Distribucin de clases")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")
plt.show()
```
